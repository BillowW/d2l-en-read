```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
alphas = [0, 0.2, 0.4, .6, .8, 1]
```

```python
x = torch.arange(-2, 1, 0.1)
```

```python
Y = [nn.LeakyReLU(alpha)(x).numpy() for alpha in alphas]
```

```python
Y
```

```python
d2l.plot(x.numpy(), Y, 'x', 'y', alphas)
```
